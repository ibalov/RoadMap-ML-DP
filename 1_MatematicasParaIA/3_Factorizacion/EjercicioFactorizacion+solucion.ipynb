### Ejercicios

Basándonos en los contenidos vistos en la clase teórica, usaremos las representaciones matriciales de las transfomaciones lineales pero aplicadas a $R^3$.

E.1. Generar una transformación lineal sobre el espacio vectorial de $R^3$ representada por una matriz que implique i) una expansión con un escalar de 2 en el eje 'y' y un escalar de 3 en el eje 'z' seguida de ii) una proyección en el eje 'x' e 'y'. Calcular los autovalores y autovectores (utilizando las funciones vistas).

E.2. Generar una transformación lineal sobre el espacio vectorial de $R^3$ representada por una matriz que implique i) una proyección sobre los ejes x e y seguida de ii) una rotación de 60 grados en el plano xz. Calcular los autovalores y autovectores (utilizando las funciones vistas).

Utilizar las siguientes matrices de rotación en $R^3$ en caso de neceistarlas

$$M_{yz} = \begin{bmatrix}1 &  0 &  0\\0 & cos(\phi) & -sen(\phi)\\ 0 &  sen(\phi) & cos(\phi) \end{bmatrix}$$

$$M_{xz} = \begin{bmatrix}cos(\phi) &  0 &  sen(\phi)\\0 & 1 & 0\\ -sen(\phi) & 0 & cos(\phi) \end{bmatrix}$$

$$M_{xy} = \begin{bmatrix}cos(\phi) &  -sen(\phi) &  0\\sen(\phi) & cos(\phi) & 0\\ 0 &  0 & 1 \end{bmatrix}$$

3. Graficar la acción de las trasnformaciones lineales de los puntos anteriores sobre cada vector unitrio con componente en 'x', 'y' y 'z'

4. Dada la siguiente matriz:
$$M = \begin{bmatrix}-0.27 &  1.36 &  0.90\\0.72 & -2.63 & -1.09\\ 0.72 &  0.36 & -0.75 \end{bmatrix}$$

Hallar los autovectores y autovalores utilizando las funciones vistas en la clase teórica y construir su representación diagonal (usando la matriz de autovalores y las matrices de autovectores $X$ y $X^{-1}$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1
E_y=np.matrix([[1,0,0],[0,2,0],[0,0,1]]) #Generamos las matrices de expansión según los escalares indicados ubicándolos en la componenete especificada
E_z=np.matrix([[0,0,0],[0,1,0],[0,0,3]]) #Generamos las matrices de expansión según los escalares indicados ubicándolos en la componenete especificada
P_xy=np.matrix([[1,0,0],[0,1,0],[0,0,0]]) #Generamos las matriz de proyección sobre los ejes 'x' e 'y' dejando la componente en 'z' multiplicada por cero

M_1 = (P_xy @ E_z @ E_y) # Generamos la matriz asociada a la trasnformación lineal como el producto amtricial de las transformaciones generadas

print("Matriz de la Transformaciòn pedida")

print(M_1)

print("Autovalores y autovectores")

eigenvalues, eigenvectors = np.linalg.eig(M_1) #calculamos autovectores y autovalres según lo pedido
print(eigenvalues[0], eigenvectors.T[0])
print(eigenvalues[1], eigenvectors.T[1])
print(eigenvalues[2], eigenvectors.T[2])

```

```python
#2.

ang=3.14159/3 # Definimos el àngulo de 60 grados para la rotación pedida
d_1=np.cos(ang) # Calculamos los valores de las funciones trigonomètricas a utilizar en la matriz de rotación
d_2=np.sin(ang) # Calculamos los valores de las funciones trigonomètricas a utilizar en la matriz de rotación
d_3=-np.sin(ang) # Calculamos los valores de las funciones trigonomètricas a utilizar en la matriz de rotación

R_xy=np.matrix([[d_1,d_3,0],[d_2,d_1,0],[0,0,1]]) # Calculamos las matrices de rotación en R3
R_xz=np.matrix([[d_1,0,d_2],[0,1,0],[d_3,0,d_1]])
R_yz=np.matrix([[1,0,0],[0,d_1,d_3],[0,d_2,d_1]])

P_xy=np.matrix([[1,0,0],[0,1,0],[0,0,0]]) # Calculamos las matrices de proyecciòn en R3 sobre cada plano
P_xz=np.matrix([[1,0,0],[0,0,0],[0,0,1]])
P_yz=np.matrix([[0,0,0],[0,1,0],[0,0,1]])


M_2 = (R_xz @ P_xy ) # Generamos la matriz asociada a la trasnformación lineal como el producto amtricial de las transformaciones generadas

print("Matriz de la Transformaciòn pedida")

print(M_2)

print("Autovalores y autovectores")

eigenvalues, eigenvectors = np.linalg.eig(M_2) #calculamos autovectores y autovalres según lo pedido
print(eigenvalues[0], eigenvectors.T[0])
print(eigenvalues[1], eigenvectors.T[1])
print(eigenvalues[2], eigenvectors.T[2])
```

```python
#3.A
v1=[1,0,0] # Definimos vector unitario en el eje correspondiente
v2_1=np.array((M_1 @ v1))[0] # Generamos la transformaciòn para el vector dado para las dos transformaciones calculadas
v2_2=np.array((M_2 @ v1))[0] # Generamos la transformaciòn para el vector dado para las dos transformaciones calculadas
fig = plt.figure() # Ploteamos
ax = plt.axes(projection='3d')
ax.plot3D([0,v1[0]],[0,v1[1]],[0,v1[2]], 'blue')
ax.plot3D([0,v2_1[0]],[0,v2_1[1]],[0,v2_1[2]], 'red')
ax.plot3D([0,v2_2[0]],[0,v2_2[1]],[0,v2_2[2]], 'green')
```

```python
#3.B
v1=[0,1,0] # Definimos vector unitario en el eje correspondiente
v2_1=np.array((M_1 @ v1))[0] # Generamos la transformaciòn para el vector dado para las dos transformaciones calculadas
v2_2=np.array((M_2 @ v1))[0] # Generamos la transformaciòn para el vector dado para las dos transformaciones calculadas
fig = plt.figure() # Ploteamos
ax = plt.axes(projection='3d')
ax.plot3D([0,v1[0]],[0,v1[1]],[0,v1[2]], 'blue')
ax.plot3D([0,v2_1[0]],[0,v2_1[1]],[0,v2_1[2]], 'red')
ax.plot3D([0,v2_2[0]],[0,v2_2[1]],[0,v2_2[2]], 'green')
```
